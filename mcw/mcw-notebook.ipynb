#### Імпорт необхідних бібліотек

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

#### Визначення параметрів моделі

```python
A = 0.5            # Амплітуда
f = 1.0            # Частота в Гц
phi = 0            # Фаза у радіанах
sigma = 0.2        # Стандартне відхилення похибки
t_end = 8          # Кінцевий час
num_points = 500   # Кількість точок
```

#### Створення періодичної моделі та нормального шуму

```python
t = np.linspace(0, t_end, num_points)
omega = 2 * np.pi * f
signal = A * np.sin(omega * t + phi)
noise = np.random.normal(0, sigma, num_points)
measurement = signal + noise
```

#### Гістограма розподілу похибок

```python
plt.figure(figsize=(12, 6))
plt.hist(noise, bins=30, density=True, alpha=0.5, label="Розподіл похибок")
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 0, sigma)
plt.plot(x, p, "k", linewidth=2, label="Теоретична нормальна крива")
plt.xlabel("Похибка")
plt.ylabel("Ймовірність")
plt.title("Гістограма розподілу похибок")
plt.legend()
plt.grid(True)
plt.show()
```

#### Графік моделі та вимірювань з похибкою

```python
plt.figure(figsize=(12, 6))
plt.plot(t, measurement, label="Вимірювання з похибкою", color="orange")
plt.plot(t, signal, label="Модель даних")
plt.xlabel("Час")
plt.ylabel("Амплітуда")
plt.title("Модель виміру за періодичним законом з нормальним законом похибки")
plt.legend()
plt.grid(True)
plt.show()
```
