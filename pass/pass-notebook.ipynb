#### Імпорт необхідних бібліотек

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

#### Визначення параметрів моделі `y = ax^2 + bx + c`

```python
n = 400       # кількість точок (вимірювань)
a = 2.0       # коефіцієнт a
b = 3.0       # коефіцієнт b
c = -1.0      # коефіцієнт c
sigma = 10.0  # стандартне відхилення похибки
```

#### Створення квадратичної моделі та нормального шуму

```python
x_t = np.linspace(-10, 10, n)
y_t = a * x_t**2 + b * x_t + c
noise = np.random.normal(0, sigma, size=n)
measurement = y_t + noise
```

#### Гістограма розподілу похибок

```python
plt.figure(figsize=(12, 6))
plt.hist(noise, bins=30, density=True, alpha=0.5, label="Розподіл похибок")
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 0, sigma)
plt.plot(x, p, "k", linewidth=2, label="Теоретична нормальна крива")
plt.xlabel("Похибка")
plt.ylabel("Ймовірність")
plt.title("Гістограма розподілу похибок")
plt.legend()
plt.grid(True)
plt.show()
```

#### Графік моделі та вимірювань з похибкою

```python
plt.figure(figsize=(12, 6))
plt.plot(x_t, measurement, label="Вимірювання з похибкою", color="orange")
plt.plot(x_t, y_t, label="Квадратична модель даних")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Модель виміру за квадратичним законом з нормальним розподілом похибок")
plt.legend()
plt.grid(True)
plt.show()
```
